```python
df = pd.read_excel('nutrient.xls', sheet_name='Nutrients',na_filter=False)
# Find the index of rows where the column is 'Infants'
infant_indexes = df[df['NUTRIENT REFERENCE VALUES FOR AUSTRALIA AND NEW ZEALAND'] == 'Infants'].index

# Create a list of indexes to drop, which includes the index found as well as one before and after
indexes_to_drop = set()
for idx in infant_indexes:
    indexes_to_drop.update({idx-1, idx, idx+1})

# Drop the rows by index
clean_df = df.drop(indexes_to_drop)
new_header = clean_df.iloc[1]
clean_df = clean_df[2:]
clean_df.columns = new_header
clean_df.reset_index(drop=True, inplace=True)

prep_df = clean_df.drop(index=[0,1, 2, 3])
prep_df.reset_index(drop=True, inplace=True)
```

```python
prep_df['Gender'] = prep_df['Gender'].replace('', 'Apply for male and female')
prep_df['Age Range_upper'] = prep_df['Age Range_upper'].replace('', 'No limit')

prep_df["Age Range_lower"] = prep_df["Age Range_lower"].apply(lambda x: int(x.replace('y', '')))
prep_df["Age Range_upper"] = prep_df["Age Range_upper"].apply(lambda x: int(x.replace('y', '')) if x.replace('y', '').isdigit() else x)
```

```python
 def process_nutrient_values_adjusted(row):
    """
    Process the nutrient values in a row to handle ranges, EAR/RDI/UL, and special texts.
    Adjusts to include terms before the second underscore for "Dietary fats" columns.
    Returns a dictionary with structured information for each nutrient.
    """
    structured_data = {}

    for col in row.index:
        value = row[col]
        # Adjusting for "Dietary fats" to include the term before the second underscore
        if "Dietary fats" in col:
            nutrient = "_".join(col.split('_')[:2])  # Include term before second underscore for Dietary fats
        else:
            nutrient = col.split('_')[0]  # Original approach for other nutrients
        
        measurement = col.split('_')[-1]

        if nutrient not in structured_data:
            structured_data[nutrient] = {}

        # Handling ranges
        if '–' in str(value):
            min_val, max_val = value.split('–')
            structured_data[nutrient]['range'] = [float(min_val), float(max_val)]

        # Handling special texts
        elif value == 'NP':
            structured_data[nutrient][measurement] = "Not possible to set upper limit"
        elif value == 'ND':
            structured_data[nutrient][measurement] = "Not determined"
        else:
            try:
                # Handling numeric values including EAR, RDI, UL
                structured_data[nutrient][measurement] = float(value)
            except ValueError:
                # In case of unexpected format, record as is
                structured_data[nutrient][measurement] = value

    return structured_data

# Re-process the example row with the adjusted function
example_row = prep_df.loc[prep_df['Age Range_lower'] <= 5].iloc[0]
processed_nutrients_adjusted = process_nutrient_values_adjusted(example_row)
processed_nutrients_adjusted  # Display some of the processed data for inspection
[400,500]
```
